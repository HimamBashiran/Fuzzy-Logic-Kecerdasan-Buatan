## Penjelasan mengenai dua metode yang digunakan

Metode yang Digunakan
1. Metode Mamdani:
- Metode Mamdani merupakan salah satu metode yang menawarkan pendekatan adaptif untuk mengatasi ketidakpastian dan kesamaran data dalam pengambilan keputusan (Muhammad Rizky Tri Nugroho). Metode ini menghasilkan output berupa fungsi keanggotaan fuzzy, yang kemudian diubah menjadi nilai numerik (crisp) melalui proses defuzzifikasi.
Metode Mamdani digunakan untuk **mengklasifikasikan wilayah berdasarkan timbulan sampah tahunan dan luas wilayah**. Berikut langkahnya:

    1. **Definisi Input dan Output:**
- Input: Timbulan sampah tahunan (`waste_volume`) dan luas wilayah (`area_size`), dengan kategori fuzzy (low, medium, high).
- Output: Kategori wilayah (`region_category`), juga dalam kategori fuzzy (low, moderate, high).

    2. **Fungsi Keanggotaan:**  
   Fungsi keanggotaan berbentuk trapesium digunakan untuk input dan output, memetakan nilai numerik (crisp) menjadi derajat keanggotaan fuzzy.

    3. **Aturan Fuzzy:**  
Terdapat 9 aturan berbasis logika **IF-THEN** yang menghubungkan input dengan output, seperti:  

    4. **Inferensi Fuzzy:**  
   Aturan diterapkan menggunakan operator **AND** (min), menghasilkan fungsi keanggotaan fuzzy untuk output.

    5. **Defuzzifikasi:**  
   Output fuzzy digabungkan dan dihitung nilai crisp-nya menggunakan metode **centroid**, memberikan hasil akhir numerik untuk kategori wilayah.

2. Metode Sugeno:
- Metode TakagiSugeno merupakan metode inferensi fuzzy untuk aturan yang direpresentasikan dalam bentuk (IF – THEN) dimana output sistem tidak berupa himpunan fuzzy, melainkan berupa konstanta atau persamaan linear (Darma Yunita Darmawi). Metode Sugeno digunakan untuk mengklasifikasikan wilayah berdasarkan timbulan sampah tahunan dan luas wilayah. Berikut langkahnya:

    1. **Definisi Input dan Output:**  
- Input: Timbulan sampah tahunan (`waste_volume`) dan luas wilayah (`area_size`), dengan kategori fuzzy (low, medium, high).  
- Output: Kategori wilayah (`region_category`), berupa nilai konstan (30, 60, 90) yang ditentukan oleh aturan.  

   2. **Fungsi Keanggotaan:**  
Fungsi keanggotaan berbentuk trapesium digunakan untuk input, memetakan nilai numerik (crisp) menjadi derajat keanggotaan fuzzy. Output pada Sugeno tidak menggunakan fungsi keanggotaan, melainkan nilai konstan.  

   3. **Aturan Fuzzy:**  
Terdapat 9 aturan berbasis logika **IF-THEN** yang menghubungkan input dengan output. Contoh:  
- Jika waste_volume rendah dan area_size kecil, maka region_category = 30.
- Jika waste_volume tinggi dan area_size besar, maka region_category = 90.  

   4. **Inferensi Fuzzy:**  
- Aturan diterapkan menggunakan operator **AND** (min) untuk menghitung tingkat kecocokan input dengan aturan.  
- Setiap aturan menghasilkan nilai output konstan yang dibobotkan oleh tingkat kecocokan (derajat keanggotaan).  

   5. **Agregasi dan Defuzzifikasi:**  
- Nilai output dihitung sebagai **rata-rata berbobot** (weighted average) dari semua nilai konstan yang relevan:  
  \[
  \text{Output Sugeno} = \frac{\sum (\text{derajat keanggotaan aturan} \times \text{nilai konstan output})}{\sum (\text{derajat keanggotaan aturan})}
  \]  
- Hasil akhirnya adalah nilai crisp yang mewakili kategori wilayah.

Dengan metode Sugeno, output berupa nilai numerik yang presisi dan komputasinya lebih sederhana dibandingkan Mamdani. Output ini cocok untuk sistem yang membutuhkan efisiensi tinggi, seperti pengendalian otomatis atau pengambilan keputusan berbasis data.

### Desain Sistem Fuzzy untuk Klasifikasi Volume Sampah di Kabupaten/Kota di Indonesia Tahun 2023 dengan Metode Mamdani

#### Fuzzifikasi

1. Menentukan input dan output:
    - Berdasarkan dataset ini, maka inputnya adalah timbulan sampah tahunan (ton) dan luas wilayah (km²), sedangkan untuk outputnya adalah kategori volume sampah di wilayah tersebut.
    - Desain linguistic variabel untuk setiap input:
        1. timbulan sampah tahunan (ton): low, medium, high
        2. luas wilayah (km²): small, medium, large
        3. Untuk output, kami desain juga 3 linguistic (Wilayah = High, Moderate, Low)

###### Definisikan Aturan

Sebelumnya kami mendefinisikan sebuah aturan, sesuai pernyataan dari Prof Anditya Arifianto bahwa idealnya untuk sebuah aturan atau fungsi itu didefinisikan atau disediakan oleh ahli/pakar di bidang tersebut. Jadi, aturan yang kami buat dibawah ini merupakan aturan yang belum tentu benar.

```python
waste_volume = ctrl.Antecedent(np.arange(0, 200001, 1000), 'waste_volume')  # Timbulan sampah tahunan (ton)
area_size = ctrl.Antecedent(np.arange(0, 31000, 100), 'area_size')          # Luas wilayah (km2)
```

Mendefinisikan aturan input fuzzy, yaitu:
   - waste_volume mewakili timbulan sampah tahunan dalam rentang 0–200.000 ton, dengan langkah 1.000 ton.
   - area_size mewakili luas wilayah dalam rentang 0–31.000 km², dengan langkah 100 km².

##### Definisikan Output

```python
region_category = ctrl.Consequent(np.arange(0, 101, 1), 'region_category')  # Kategori wilayah
```

Mendefinisikan aturan output fuzzy, yaitu: region_category dengan rentang nilai 0–100 dan langkah 1. 
Variabel ini akan digunakan untuk menentukan kategori wilayah berdasarkan inferensi fuzzy dari input, seperti low, moderate, dan high.

##### Definisikan Bentuk dan Batas Fungsi Keanggotaan (Membership Functions)

```python
waste_volume['low'] = fuzz.trapmf(waste_volume.universe, [0, 0, 20000, 40000])
waste_volume['medium'] = fuzz.trapmf(waste_volume.universe, [30000, 50000, 70000, 90000])
waste_volume['high'] = fuzz.trapmf(waste_volume.universe, [80000, 120000, 200000, 200000])
```

Mendefinisikan fungsi keanggotaan fuzzy untuk variabel input waste_volume (timbulan sampah tahunan) menggunakan fungsi trapesium (trapmf):
- waste_volume['low']:  Timbulan sampah dianggap rendah jika berada di bawah 20,000 ton, dan mulai kurang "rendah" saat mendekati 40,000 ton.
- waste_volume['medium']: Timbulan sampah dianggap sedang jika berada di sekitar 50,000–70,000 ton, dan mulai kurang "sedang" di luar rentang tersebut.
- waste_volume['high']: Timbulan sampah dianggap tinggi jika di atas 120,000 ton, dan mulai kurang "tinggi" jika mendekati 80,000 ton.

```python
area_size['small'] = fuzz.trapmf(area_size.universe, [0, 0, 2000, 4000])
area_size['medium'] = fuzz.trapmf(area_size.universe, [3000, 5000, 7000, 9000])
area_size['large'] = fuzz.trapmf(area_size.universe, [8000, 15000, 30000, 30000])
```

Mendefinisikan fungsi keanggotaan fuzzy untuk variabel input area_size (ukuran area) menggunakan fungsi trapesium (trapmf):
- area_size['small']: Ukuran area dianggap kecil jika berada di bawah 2,000 m², dan mulai kurang "kecil" saat mendekati 4,000 m².
- area_size['medium']: Ukuran area dianggap sedang jika berada di sekitar 3,000–5,000 m², dan mulai kurang "sedang" di luar rentang tersebut.
- area_size['large']: Ukuran area dianggap besar jika di atas 8,000 m², dan mulai kurang "besar" jika mendekati 15,000 m².

```python
region_category['low'] = fuzz.trapmf(region_category.universe, [0, 0, 30, 50])
region_category['moderate'] = fuzz.trapmf(region_category.universe, [40, 50, 70, 90])
region_category['high'] = fuzz.trapmf(region_category.universe, [80, 90, 100, 100])
```

Mendefinisikan fungsi keanggotaan fuzzy untuk variabel output region_category (kategori wilayah) menggunakan fungsi trapesium (trapmf):
- region_category['low']: Kategori wilayah dianggap rendah jika nilainya antara 0–30, dan mulai kurang "rendah" hingga 50.
- region_category['moderate']: Kategori wilayah dianggap sedang jika nilainya antara 50–70, dan mulai kurang "sedang" di luar rentang 40–90.
- region_category['high']: Kategori wilayah dianggap tinggi jika nilainya antara 90–100, dan mulai kurang "tinggi" di bawah 80.

Fungsi ini digunakan untuk menentukan seberapa besar wilayah tersebut masuk ke dalam kategori low, moderate, atau high berdasarkan hasil inferensi fuzzy.

#### Inference

Fuzzy Inference System (FIS) atau biasa dikenal dengan Inference merupakan sebuah teknik pemetaan sebuah ruang input menjadi sebuah ruang output yang berbeda menggunakan logika fuzzy (Juju Jumadi). Proses ini menerapkan aturan untuk mengubah input fuzzy menjadi output fuzzy, yang kemudian dapat didefuzzifikasi untuk menghasilkan keputusan atau nilai akhir.

##### Mendefinisikan Aturan Fuzzy

```python
rule1 = ctrl.Rule(waste_volume['low'] & area_size['small'], region_category['low'])
rule2 = ctrl.Rule(waste_volume['low'] & area_size['medium'], region_category['moderate'])
rule3 = ctrl.Rule(waste_volume['low'] & area_size['large'], region_category['high'])
rule4 = ctrl.Rule(waste_volume['medium'] & area_size['small'], region_category['moderate'])
rule5 = ctrl.Rule(waste_volume['medium'] & area_size['medium'], region_category['moderate'])
rule6 = ctrl.Rule(waste_volume['medium'] & area_size['large'], region_category['high'])
rule7 = ctrl.Rule(waste_volume['high'] & area_size['small'], region_category['high'])
rule8 = ctrl.Rule(waste_volume['high'] & area_size['medium'], region_category['high'])
rule9 = ctrl.Rule(waste_volume['high'] & area_size['large'], region_category['high'])
```

Mendefinisikan aturan-aturan fuzzy untuk sistem kontrol berbasis logika fuzzy menggunakan variabel input waste_volume (timbulan sampah tahunan) dan area_size (ukuran area), serta variabel output region_category (kategori wilayah). Aturan-aturan ini menentukan hubungan antara kategori volume sampah dan ukuran area dengan kategori wilayah.

| Aturan | waste_volume | area_size | region_category |
|--------|--------------|-----------|-----------------|
| 1      | rendah       | kecil     | rendah          |
| 2      | rendah       | sedang    | sedang          |
| 3      | rendah       | besar     | tinggi          |
| 4      | sedang       | kecil     | sedang          |
| 5      | sedang       | sedang    | sedang          |
| 6      | sedang       | besar     | tinggi          |
| 7      | tinggi       | kecil     | tinggi          |
| 8      | tinggi       | sedang    | tinggi          |
| 9      | tinggi       | besar     | tinggi          |

Secara keseluruhan, aturan-aturan ini menggambarkan bagaimana sistem fuzzy dapat memetakan kombinasi input ke dalam kategori output tertentu.

```python
region_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
```

Mendefinisikan sistem kontrol fuzzy berdasarkan aturan-aturan fuzzy yang telah ditentukan sebelumnya

```python
kategori_wilayah = ctrl.ControlSystemSimulation(region_ctrl)
```

### Desain Sistem Fuzzy untuk Klasifikasi Volume Sampah di Kabupaten/Kota di Indonesia Tahun 2023 dengan Metode Takagi - Sugeno Style

Selanjutnya, kami menganalisis klasifikasi volume sampah di kabupaten/kota di Indonesia tahun 2023 dengan metode sugeno. Berbeda dengan Mamdani yang menghasilkan output fuzzy, memerlukan defuzzifikasi dengan centroid untuk menghasilkan nilai crisp, metode Sugeno menghasilkan output numerik langsung (tanpa agregasi fuzzy) dengan rata-rata berbobot, lebih sederhana dan efisien.

Berikut adalah perbedaan antara Metode Mamdani dan Metode Sugeno:


| Aspek | Mamdani | Sugeno |
|--------|--------------|-----------|
| Output      |Fungsi keanggotaan fuzzy (low, moderate, high)| Nilai konstan (30, 60, 90)  | 
| Penerapan Aturan| Menghubungkan input fuzzy dengan output fuzzy|Menghubungkan input fuzzy dengan output konstan| 
| Agregasi|Output fuzzy digabungkan menjadi satu fungsi fuzzy|Nilai output dijumlahkan dengan weighted average| 
| Deffuzifikasi| 	Diperlukan, menggunakan metode centroid|Tidak diperlukan, hasil langsung dihitung| 
| Kompleksitas|	Lebih kompleks karena memproses fuzzy output dan defuzzifikasi|Lebih sederhana karena langsung menghasilkan nilai crisp| 

```python
# Definisikan input rentang nilai
waste_universe = np.arange(0, 200001, 1000)  # Timbulan sampah tahunan (ton)
area_universe = np.arange(0, 31000, 100)    # Luas wilayah (km2)
```

Mendefinisikan rentang nilai (universe of discourse) dari setiap variabel input yang akan digunakan dalam sistem fuzzy:
- waste_universe: Rentang nilai untuk timbulan sampah tahunan, dari 0 hingga 200.000 ton, dengan interval 1.000 ton.
- area_universe: Rentang nilai untuk luas wilayah, dari 0 hingga 31.000 km², dengan interval 100 km².

```python
# Definiskan fungsi keanggotaan untuk waste_volume
waste_low = fuzz.trapmf(waste_universe, [0, 0, 20000, 40000])
waste_medium = fuzz.trapmf(waste_universe, [30000, 50000, 70000, 90000])
waste_high = fuzz.trapmf(waste_universe, [80000, 120000, 200000, 200000])
```

Mendefinisikan fungsi keanggotaan fuzzy untuk variabel input waste_volume (timbulan sampah tahunan) menggunakan bentuk trapesium.
- waste_low: Timbulan sampah dianggap rendah jika berada dalam rentang 0 hingga 40.000 ton. Menurun secara linear dari 20.000 hingga 40.000 ton.
- waste_medium: Timbulan sampah dianggap sedang jika berada dalam rentang 30.000 hingga 90.000 ton. Menurun secara linear di luar rentang tersebut.
- waste_high: Timbulan sampah dianggap tinggi jika berada dalam rentang 80.000 hingga 200.000 ton. Menurun secara linear dari 80.000 hingga 120.000 ton.

```python
# Definiskan fungsi keanggotaan untuk area_size
area_small = fuzz.trapmf(area_universe, [0, 0, 2000, 4000])
area_medium = fuzz.trapmf(area_universe, [3000, 5000, 7000, 9000])
area_large = fuzz.trapmf(area_universe, [8000, 15000, 30000, 30000])
```

Mendefinisikan fungsi keanggotaan fuzzy untuk variabel input area_size (luas wilayah) menggunakan bentuk trapesium:
- area_small: Wilayah dianggap kecil jika luasnya berada di bawah 2.000 km² dan mulai kurang "kecil" hingga mencapai 4.000 km².
- area_medium: Wilayah dianggap sedang jika luasnya berada di rentang 5.000–7.000 km² dan mulai kurang "sedang" jika luasnya di luar rentang 3.000–9.000 km².
- area_large: Wilayah dianggap besar jika luasnya di atas 15.000 km² dan mulai kurang "besar" jika luasnya mendekati 8.000 km².

```python
# Aturan sugeno dengan nilai konstan
def sugeno_inference(waste, area):
    """Custom Sugeno inference logic."""
    # Pastikan input values dari tiap universe
    waste = np.clip(waste, waste_universe.min(), waste_universe.max())
    area = np.clip(area, area_universe.min(), area_universe.max())
    
    # Menghitung derajat keanggotaan
    waste_low_degree = fuzz.interp_membership(waste_universe, waste_low, waste)
    waste_medium_degree = fuzz.interp_membership(waste_universe, waste_medium, waste)
    waste_high_degree = fuzz.interp_membership(waste_universe, waste_high, waste)
    
    area_small_degree = fuzz.interp_membership(area_universe, area_small, area)
    area_medium_degree = fuzz.interp_membership(area_universe, area_medium, area)
    area_large_degree = fuzz.interp_membership(area_universe, area_large, area)
    
    # Aturan sugeno berupa nilai konstan
    rule1 = min(waste_low_degree, area_small_degree) * 30
    rule2 = min(waste_low_degree, area_medium_degree) * 60
    rule3 = min(waste_low_degree, area_large_degree) * 90
    rule4 = min(waste_medium_degree, area_small_degree) * 60
    rule5 = min(waste_medium_degree, area_medium_degree) * 60
    rule6 = min(waste_medium_degree, area_large_degree) * 90
    rule7 = min(waste_high_degree, area_small_degree) * 90
    rule8 = min(waste_high_degree, area_medium_degree) * 90
    rule9 = min(waste_high_degree, area_large_degree) * 90
    
    # Terapkan Aturan Agregasi
    weighted_sum = rule1 + rule2 + rule3 + rule4 + rule5 + rule6 + rule7 + rule8 + rule9
    total_weight = (
        waste_low_degree * area_small_degree +
        waste_low_degree * area_medium_degree +
        waste_low_degree * area_large_degree +
        waste_medium_degree * area_small_degree +
        waste_medium_degree * area_medium_degree +
        waste_medium_degree * area_large_degree +
        waste_high_degree * area_small_degree +
        waste_high_degree * area_medium_degree +
        waste_high_degree * area_large_degree
    )
    
    # Defuzzifikasi (Weighted Average)
    return weighted_sum / total_weight if total_weight != 0 else 0
```

Membuat Sugeno Inference Logic dengan aturan, perhitungan derajat keanggotaan, aturan sugeno dengan nilai konstan, agregasi, dan defuzzifikasi. 
1. Aturan Fuzzy Sugeno dengan Nilai Konstan

Aturan Sugeno dalam kode ini menggunakan logika IF-THEN untuk memetakan hubungan antara input (waste_volume dan area_size) dengan output (region_category) yang berbentuk nilai konstan.
Contohnya:

    Rule 1
- Jika waste_volume rendah dan area_size kecil, maka region_category = 30.

    Rule 2
- Jika waste_volume rendah dan area_size sedang, maka region_category = 60.

2. Menghitung Derajat Keanggotaan (Membership Degree)

Menghitung derajat keanggotaan untuk setiap input berdasarkan fungsi keanggotaan trapesium yang didefinisikan sebelumnya.

3. Evaluasi Aturan dengan Operator AND (Min)

Untuk setiap aturan, operator AND digunakan untuk menghitung derajat keanggotaan dengan mengambil nilai minimum antara input fuzzy.

4. Agregasi Aturan

Semua aturan yang relevan dihitung dan dijumlahkan menggunakan rata-rata berbobot (weighted average).

5. Defuzzifikasi (Weighted Average)

Hasil akhir dihitung menggunakan formula rata-rata berbobot: 
Crisp Output= Weight Sum / Total Weighted


```python
daerah_list = df['kabupaten_kota'].tolist()

for index, row in df.iterrows():
    waste = row['timbulan_sampah_tahunan']
    area = row['luas_wilayah']
    
    # Compute Sugeno output
    category = sugeno_inference(waste, area)
    
    # Output results
    daerah = daerah_list[index % len(daerah_list)]
    print(f"{index + 1} - Daerah: {daerah}, Timbulan Sampah: {waste}, Luas Wilayah: {area}")
    print(f"Kategori Wilayah: {category:.2f}")
    print(" ")
```

Hasil konversi tersebut merupakan aksi yang diambil oleh sistem kendali logika fuzzy (Sutikno). Simpelnya, mengubah kembali nilai fuzzy menjadi nilai crisp.
